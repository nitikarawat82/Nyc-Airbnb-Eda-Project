```python
#heat map - correlation of one variable with others for numerical column, we will use all number columns as correlation works only on the number columns

corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr() #creating corr
corr
```

```python
#Lets cerate a heatmap for this
plt.figure(figsize=(8, 6))
sns.heatmap(data=corr, annot=True)
```

### The correlation heatmap shows that most features have weak relationships. However, number_of_reviews and reviews_per_month have a strong positive correlation (0.63), indicating that frequently reviewed listings receive more monthly reviews. Additionally, price shows a moderate positive correlation with beds (0.42), suggesting that larger accommodations tend to be more expensive.
Number_of_reviews ↔ reviews_per_month = 0.63 (Strongest relationship), 
 More beds → higher price (Moderate positive correlation)
